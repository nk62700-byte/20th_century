```python
import pandas as pd
```

```python
data = pd.DataFrame({"Name": ["Anna", "John", "Maria", "Leo", "Sophie"],
    "Score": [78, 85, 92, 69, 88]})
```

```python
data
```

```python
# division
data['Score_div_10'] = data['Score'].apply(lambda x: x / 10)
```

```python
data
```

```python
# multiplication
data['Score_x2'] = data['Score'].apply(lambda x: x * 2)
```

```python
data
```

```python
# addition
data['Score_plus_5'] = data['Score'].apply(lambda x: x + 5)
```

```python
data
```

```python
data['starts_with_A'] = data['Name'].apply(lambda x: x.startswith('A'))
```

```python
data
```
